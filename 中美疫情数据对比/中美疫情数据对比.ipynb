```python
import  pandas as pd
import plotly.graph_objects as go
cconfirmed={}
uconfirmed={}
start=[22,1,1,1,1]
end=[31,29,31,30,25]
for month in range(1,6):
    for day in range(start[month-1],end[month-1]+1):
        if day<10:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-0'+str(day)+'-2020.csv')
        else:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-'+str(day)+'-2020.csv')
        if month<3: 
              res=dict(data.groupby("Country/Region")["Confirmed"].sum())
        elif month==3:
            if day<=21:
                res=dict(data.groupby("Country/Region")["Confirmed"].sum())
                
            else:
                res=dict(data.groupby("Country_Region")["Confirmed"].sum())
        else:
            res=dict(data.groupby("Country_Region")["Confirmed"].sum())
        if "Mainland China" in res:
            cconfirmed[str(month)+"-"+str(day)]=res["Mainland China"]
            if "Taiwan*" in res:
                cconfirmed[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan*"]
            if "Taiwan" in res:
                cconfirmed[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan"]
        if "China" in res:
            cconfirmed[str(month)+"-"+str(day)]=res["China"]
            if "Taiwan*" in res:
                cconfirmed[str(month)+"-"+str(day)]=res["China"]+res["Taiwan*"]
            if "Taiwan" in res:
                cconfirmed[str(month)+"-"+str(day)]=res["China"]+res["Taiwan"]
        if "US" in res:
            uconfirmed[str(month)+"-"+str(day)]=res["US"]       
line1=go.Scatter(x=list(cconfirmed.keys()),y=list(cconfirmed.values()),name="中国累计确诊")
line2=go.Scatter(x=list(uconfirmed.keys()),y=list(uconfirmed.values()),name="美国累计确诊")
cdeaths={}
udeaths={}
for month in range(1,6):
    for day in range(start[month-1],end[month-1]+1):
        if day<10:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-0'+str(day)+'-2020.csv')
        else:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-'+str(day)+'-2020.csv')
        if month<3: 
              res=dict(data.groupby("Country/Region")["Deaths"].sum())
        elif month==3:
            if day<=21:
                res=dict(data.groupby("Country/Region")["Deaths"].sum())
                
            else:
                res=dict(data.groupby("Country_Region")["Deaths"].sum())
        else:
            res=dict(data.groupby("Country_Region")["Deaths"].sum())
        if "Mainland China" in res:
            cdeaths[str(month)+"-"+str(day)]=res["Mainland China"]
            if "Taiwan*" in res:
                cdeaths[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan*"]
            if "Taiwan" in res:
                cdeaths[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan"]
        if "China" in res:
            cdeaths[str(month)+"-"+str(day)]=res["China"]
            if "Taiwan*" in res:
                cdeaths[str(month)+"-"+str(day)]=res["China"]+res["Taiwan*"]
            if "Taiwan" in res:
                cdeaths[str(month)+"-"+str(day)]=res["China"]+res["Taiwan"]
        if "US" in res:
            udeaths[str(month)+"-"+str(day)]=res["US"]
            
line3=go.Scatter(x=list(cdeaths.keys()),y=list(cdeaths.values()),name="中国累计死亡")
line4=go.Scatter(x=list(udeaths.keys()),y=list(udeaths.values()),name="美国累计死亡")

crecovered={}
urecovered={}
for month in range(1,6):
    for day in range(start[month-1],end[month-1]+1):
        if day<10:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-0'+str(day)+'-2020.csv')
        else:
              data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\0'+str(month)+'-'+str(day)+'-2020.csv')
        if month<3: 
              res=dict(data.groupby("Country/Region")["Recovered"].sum())
        elif month==3:
            if day<=21:
                res=dict(data.groupby("Country/Region")["Recovered"].sum())
                
            else:
                res=dict(data.groupby("Country_Region")["Recovered"].sum())
        else:
            res=dict(data.groupby("Country_Region")["Recovered"].sum())
        if "Mainland China" in res:
            crecovered[str(month)+"-"+str(day)]=res["Mainland China"]
            if "Taiwan*" in res:
                crecovered[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan*"]
            if "Taiwan" in res:
                crecovered[str(month)+"-"+str(day)]=res["Mainland China"]+res["Taiwan"]
        if "China" in res:
            crecovered[str(month)+"-"+str(day)]=res["China"]
            if "Taiwan*" in res:
                crecovered[str(month)+"-"+str(day)]=res["China"]+res["Taiwan*"]
            if "Taiwan" in res:
                crecovered[str(month)+"-"+str(day)]=res["China"]+res["Taiwan"]
        if "US" in res:
            urecovered[str(month)+"-"+str(day)]=res["US"]           
    
line5=go.Scatter(x=list(crecovered.keys()),y=list(crecovered.values()),name="中国累计治愈")
line6=go.Scatter(x=list(urecovered.keys()),y=list(urecovered.values()),name="美国累计治愈")  
fig =go.Figure([line1,line2,line3,line4,line5,line6])
fig.update_layout(
          title="中美疫情数据对比曲线",
          xaxis_title="日期",
          yaxis_title="累计病例"
      )
fig.show()
```
