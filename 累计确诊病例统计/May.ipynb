```python
import pandas as pd
import plotly.graph_objects as go
for num in range(1,10):
      data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\05-0'+str(num)+'-2020.csv')
      res=dict(data.groupby("Country_Region")["Confirmed"].sum())
      line=go.Bar(x=list(res.keys()),y=list(res.values()),name="累计确诊")
      fig =go.Figure(line)
      fig.update_layout(
          title="世界各地区/国家疫情数据-05-0"+str(num),
          xaxis_title="国家/地区",
          yaxis_title="累计新增病例"
      )
      fig.show()
for num in range(10,26):
      data=pd.read_csv('F:\\COVID 19\\COVID-19\\csse_covid_19_data\\csse_covid_19_daily_reports\\05-'+str(num)+'-2020.csv',keep_default_na=0)
      res=dict(data.groupby("Country_Region")["Confirmed"].sum())
      line=go.Bar(x=list(res.keys()),y=list(res.values()),name="累计确诊")
      fig =go.Figure(line)
      fig.update_layout(
          title="世界各地区/国家疫情数据-05-"+str(num),
          xaxis_title="国家/地区",
          yaxis_title="累计新增病例"
      )
      fig.show()
```
